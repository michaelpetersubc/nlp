```python
import spacy
nlp = spacy.load("en_core_web_sm")
nlp
```

```python
#cleantext = BeautifulSoup(raw_html, "lxml").text
#cleantext = cleantext.replace("\r", "")
#cleantext
file = open("1515209.txt", "r")
cleantext = file.read()
file.close()
```

```python
processed_text = nlp(cleantext)
```

```python
sentences = list(processed_text.sents)
len(sentences)
```

```python
n = 0
for sentence in sentences:
    print(sentence)
    n += 1
    if(n > 10):
        break
```

```python
print(
    f"{'Text with Whitespace':22}"
    f"{'Is Alphanumeric?':17}"
    f"{'Is Punctuation?':18}"
    f"{'Is Stop Word?'}"
)
for token in processed_text:
    print(
        f"{str(token.text_with_ws):22}"
        f"{str(token.is_alpha):17}"
        f"{str(token.is_punct):18}"
        f"{str(token.is_stop)}"
    )
```

```python
words = [
    token.text
    for token in processed_text
    if not token.is_stop and not token.is_punct
]
```

```python
import collections
print(collections.Counter(words).most_common(5))
```
