```python
import mysql.connector
import os, sys
from dotenv import load_dotenv
load_dotenv()
print(os.environ.get('foousername'))
```

```python
mydb = mysql.connector.connect(
  host="localhost",
  user=os.environ.get('foousername'),
  password=os.environ.get('foopassword'),
  port=os.environ.get('fooport'),
  database=os.environ.get('foodatabase')
    )

print(mydb) 
```

```python
mycursor = mydb.cursor(dictionary = True)
mycursor.execute("select file_id,aid,f.blobid,b.filesize from myfiles f join myblobs b \
                 on f.blobid=b.blobid where f.type_id=20 order by file_id")
myresult = mycursor.fetchall()
#for x in myresult:
 # print(x['file_id'])
```

```python
path = '/home/peters/working_papers_archive/'
for x in myresult:
    mycursor.execute("select filecontent from myblobs where blobid="+str(x['blobid']))
    blob = mycursor.fetchone()
    filename = path + str(x['aid']) + "_" + str(x['blobid']) + ".pdf"
    f = open(filename, "wb")
    f.write(blob['filecontent'])
    print("saved " + filename)
    
```
